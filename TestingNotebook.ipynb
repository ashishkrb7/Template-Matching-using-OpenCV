```python
import numpy as np
import cv2
import json
from matplotlib import pyplot as plt
```

```python
source_img = cv2.imread("source.png")
```

```python
plt.imshow(cv2.cvtColor(source_img, cv2.COLOR_BGR2RGB))
plt.title('Source Image')
plt.show()
```

```python
(h, w) = source_img.shape[:2]
(h, w)
```

```python
coordinate = json.load(open('output.json'))
coordinate
```

```python
out_img = cv2.rectangle(source_img, (coordinate["startX"], coordinate["startY"]), (coordinate["endX"], coordinate["endY"]), (0, 0, 255), 2)
```

```python
plt.imshow(cv2.cvtColor(out_img, cv2.COLOR_BGR2RGB))
plt.title('Ouput Image')
plt.show()
```
